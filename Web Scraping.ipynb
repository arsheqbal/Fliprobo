```python
p_n=phone.find_all("span",{"class":"a-size-medium a-color-base a-text-normal"})
p_pr=phone.find_all("span",{"class":"a-price-whole"})
p_url=phone.find_all("img",{"class":"s-image"})
p_rat=phone.find_all("span",{"class":"a-icon-alt"})
    
name=[]
price=[]
img=[]
rat=[]
for i in p_n:
    name.append(i.text.replace("\n"," "))
for i in p_pr:
    price.append(i.text.replace("\n"," "))
for i in p_url:
    img.append(i['src'])
for i in p_rat:
    rat.append(i.text.replace("\n"," "))

del rat[15:]
del img[0:1]
del name[0:1]

data={"Phone Name":[],"Price":[],"Image":[],"Rating":[]}
data["Phone Name"]=name
data["Price"]=price
data["Image"]=img
data["Rating"]=rat
df1=pd.DataFrame(data)
df1
```

# National Weather Service San Fransisco:-

```python
page = requests.get('https://forecast.weather.gov/MapClick.php?lat=37.77493000000004&lon=-122.41941999999995#.YJZJp7UzbIU')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
ind = [1,3,5,7]
period = []
short_desc = []
desc = []
temp = []

s = soup.find_all('p', class_='period-name')
for info in s:
    name = info.text
    period.append(name)

t = soup.find_all('p', class_='short-desc')
for info in t:
    name = info.text
    short_desc.append(name)

q = soup.find_all('p', class_='temp temp-low')
r = soup.find_all('p', class_='temp temp-high')
for info in q:
    name = info.text
    temp.append(name)

for info in r:
    name = info.text
    temp.insert(ind[0], name)
    del ind[0]


u = soup.find_all('div', class_='col-sm-10 forecast-text')
for info in u:
    name = info.text
    desc.append(name)
```

```python
Desc = desc[0:9]
```

```python
import pandas as pd
Weather = pd.DataFrame({})
Weather['Period'] = period
Weather['Short Description'] = short_desc
Weather['Temperature'] = temp
Weather['Description'] = Desc
```
